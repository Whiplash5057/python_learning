```python
import os, PyPDF2

f = open('Contact_Email_Information.pdf', 'rb') # read binary

email_list = []

pdf_reader = PyPDF2.PdfFileReader(f)
pdf_total_pages = pdf_reader.numPages
for n in range(pdf_total_pages):
    test_string = pdf_reader.getPage(n).extractText()
    pattern = r'[\w_.-]*@[\w]*.[\w]*'
    email_list.append(re.findall(pattern, test_string))
    
print(email_list)
f.close()
```
