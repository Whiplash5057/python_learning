```python
import bs4, requests
import os
from os import walk
import re

### --- PART ONE --- ###

# res = requests.get('https://www.thegoldbugs.com/blog')
# soup = bs4.BeautifulSoup(res.text, 'lxml')
# encrypt = soup.select('.sqs-block-content pre')[0].text
# encrypt_list = encrypt.split('-----')
# decrypt_list = []
# for sentence in encrypt_list:
#     if len(sentence) > 0:
#         decrypt_list.append(sentence[0])
#     else:
#         decrypt_list.append(sentence)
# print(''.join(decrypt_list))



urls = []
for root, dirs, files in walk('.', topdown=False):
    for name in files:
        print(os.path.join(root, name))
        open_file = open(os.path.join(root, name))
        lines = open_file.read()
#         print(lines)
        urls.append(re.findall(r'(http://|https://)(www.)*(docs.|drive.)[\w\d/=]*', lines))
        print('\n')
        print('\n')
        print('\n')
        open_file.close()

print(urls)
```
